```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define the sinusoidal model function
def sinusoidal(t, A, B, C, D):
    return A * np.sin(B * t + C) + D

# Data: months of the year (0 to 11)
months = np.arange(12)

# Given data
max_temps = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
min_temps = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])

# Initial guesses for the parameters
initial_guess = [10, 2 * np.pi / 12, 0, np.mean(max_temps)]


```

```python
# Fit the max temperature data
params_max, _ = curve_fit(sinusoidal, months, max_temps, p0=initial_guess)

# Fit the min temperature data
params_min, _ = curve_fit(sinusoidal, months, min_temps, p0=initial_guess)

# Generate a range of values for plotting the fitted function
t_fit = np.linspace(0, 11, 100)

# Plot the original data and the fitted functions
plt.figure(figsize=(12, 6))

```

```python
# Plot max temperatures
plt.subplot(2, 1, 1)
plt.scatter(months, max_temps, color='red', label='Max Temperatures')
plt.plot(t_fit, sinusoidal(t_fit, *params_max), color='blue', label='Fitted Max Temp')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.title('Max Temperatures and Fitted Function')
plt.legend()

# Plot min temperatures
plt.subplot(2, 1, 2)
plt.scatter(months, min_temps, color='green', label='Min Temperatures')
plt.plot(t_fit, sinusoidal(t_fit, *params_min), color='orange', label='Fitted Min Temp')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.title('Min Temperatures and Fitted Function')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = "https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic = pd.read_csv(url)

# 1. Create a pie chart presenting the male/female proportion

# Count the number of males and females
gender_counts = titanic['sex'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['skyblue', 'salmon'])
plt.title('Male/Female Proportion on Titanic')
plt.show()

# 2. Create a scatterplot with the Fare paid and Age, differ the plot color by gender

# Create a scatter plot
plt.figure(figsize=(10, 8))
for gender in titanic['sex'].unique():
    subset = titanic[titanic['sex'] == gender]
    plt.scatter(subset['age'], subset['fare'], label=gender, alpha=0.6)

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Fare vs Age Differentiated by Gender')
plt.legend(title='Gender')
plt.grid(True)
plt.show()
```
